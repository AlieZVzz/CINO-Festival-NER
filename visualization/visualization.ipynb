```python
import matplotlib.pyplot as plt
import numpy as np

# 支持中文
plt.rcParams['font.sans-serif'] = ['SimSun']  # 用来正常显示中文标签
plt.rcParams['axes.unicode_minus'] = False  # 用来正常显示负号
```

```python
from matplotlib import rcParams

config = {
    "font.family": 'serif', # 衬线字体
    "font.size": 12, # 相当于小四大小
    "font.serif": ['SimSun'], # 宋体
    "mathtext.fontset": 'stix', # matplotlib渲染数学字体时使用的字体，和Times New Roman差别不大
    'axes.unicode_minus': False # 处理负号，即-号
}
rcParams.update(config)
legend_font = {"family" : "Times New Roman"}
```

```python
bo = np.load('../checkpoints/bo_PLM_bilstm_crf.npy', allow_pickle=True)
plt.figure()
plt.plot(bo.item()['Festival']['f1'], label='Festival')
plt.plot(bo.item()['Item']['f1'], label='Item')
plt.plot(bo.item()['Event']['f1'], label='Event')
plt.plot(bo.item()['Location']['f1'], label="Location")
legend_font = {"family" : "Times New Roman"}
plt.legend(frameon=True, prop=legend_font)
plt.xlabel("训练轮次")
plt.ylabel(u"$\mathrm{F1}$值")
plt.title(u"藏语$\mathrm{RoBERTa-BiLSTM-CRF}$训练曲线") # $\mathrm{Times \; New \; Roman1}$
plt.savefig("藏语RoBERTa-BiLSTM-CRF训练曲线.eps", dpi=300)
```
